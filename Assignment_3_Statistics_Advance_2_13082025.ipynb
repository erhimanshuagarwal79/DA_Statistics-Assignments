**Question 1: What is hypothesis testing in statistics?**

**Answer**

Hypothesis testing is a statistical method used to make inferences or draw conclusions about a population
based on a sample of data. It involves making an assumption (hypothesis) about a population parameter,
then using sample data to test this assumption. The process determines whether to reject the assumption
based on the evidence in the data.

**Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?**

**Answer**

The null hypothesis (H₀) is a statement that there is no effect or no difference, and it represents the default assumption.
The alternative hypothesis (H₁ or Ha) contradicts the null and represents what the researcher wants to prove.
For example, if we test whether a coin is fair, H₀: p = 0.5 (coin is fair) vs H₁: p ≠ 0.5 (coin is biased).

**Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.**

**Answer**

The significance level (α) is the probability of rejecting the null hypothesis when it is actually true (Type I error).
It represents the threshold for deciding whether the observed data is statistically significant. Common α values are 0.05 or 0.01.
If the p-value is less than α, we reject the null hypothesis.

**Question 4: What are Type I and Type II errors? Give examples of each.**

**Answer**

Type I error: Rejecting the null hypothesis when it is true (false positive). Example: Concluding a medicine works when it actually does not.
Type II error: Failing to reject the null hypothesis when it is false (false negative). Example: Concluding a medicine has no effect when it actually does.

**Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.**

**Answer**

Z-test: Used when the population standard deviation is known and the sample size is large (n > 30).
T-test: Used when the population standard deviation is unknown and the sample size is small (n ≤ 30).

**Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.
(Include your Python code and output in the code box below.)
Hint: Generate random number using random function.**

```python
import numpy as np
import matplotlib.pyplot as plt

n = 10
p = 0.5
size = 1000
data = np.random.binomial(n, p, size)

plt.hist(data, bins=range(n+2), edgecolor='black', align='left')
plt.xlabel('Number of successes')
plt.ylabel('Frequency')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.show()
```

```python
import numpy as np
from scipy import stats

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

mean_pop = 50
std_pop = 0.5  # assumed known
mean_sample = np.mean(sample_data)
n = len(sample_data)

z_score = (mean_sample - mean_pop) / (std_pop / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

z_score, p_value
```

**Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.**

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

data = np.random.normal(loc=50, scale=5, size=100)
mean = np.mean(data)
std_err = stats.sem(data)
confidence = 0.95
h = std_err * stats.t.ppf((1 + confidence) / 2, len(data) - 1)

ci = (mean - h, mean + h)

plt.hist(data, bins=20, edgecolor='black')
plt.axvline(ci[0], color='red', linestyle='dashed', linewidth=2)
plt.axvline(ci[1], color='red', linestyle='dashed', linewidth=2)
plt.title('Normal Distribution with 95% CI')
plt.show()

ci
```

**Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.**

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
    """
    Calculates Z-scores for the given dataset and plots a histogram of the standardized values.
    """
    mean = np.mean(data)
    std = np.std(data, ddof=0)  # population std deviation by default
    z_scores = [(x - mean) / std for x in data]

    # Plot histogram of Z-scores
    plt.hist(z_scores, bins=10, edgecolor='black')
    plt.title('Z-score Distribution')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')
    plt.axvline(0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.legend()
    plt.show()

    return z_scores

# Example usage
data_example = np.random.randint(50, 100, size=50)
z_scores_result = calculate_and_plot_zscores(data_example)

print("First 10 Z-scores:", z_scores_result[:10])
```
